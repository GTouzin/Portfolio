# Exploration de données

Lors du mois d'octobre 2016, j'ai participé à une compétition sur le site Kaggle.com dont le but était de classer une liste d'observations dans une des trois catégories suivantes: "Ghost", "Goblin" ou bien "Goul".  Cette compétition m'a permis d'appliquer les algorithmes de partitionnement dans un contexte non compétitif et d'échanger avec des participants ayant différentes formations et expériences de travail. J'ai eu beaucoup de plaisir à modéliser ces données et dans ce billet je vais décrire comment j'ai approché le problème. Plus précisément, je vais faire l'exploration des données de la compétition et dans le billet que vous pouvez trouver à l'adresse https://github.com/GTouzin/Portfolio/blob/master/R/Portfolio_GGG_Mod%C3%A8le.ipynb, je vais résumer les étapes que j'ai suivies lors de la modélisation des données.

Dans cette compétition, nous avions à notre disposition deux jeux de donnés: une pour entraîner notre modèle et l'autre pour le tester. C'est ce dernier jeu de données qui a été utilisé pour déterminer le classement lors de la compétition.

```r
# Charger les packages à utiliser
library(data.table)
library(ggplot2)
library(ggthemes)
library(corrplot)


#Lire les données 
monstres <- fread("./Kaggle/train/train.csv",stringsAsFactors = T)
monstres_test <- fread("./Kaggle/test/test.csv",stringsAsFactors = T)
head(monstres)
#head(monstres_test)
```

De cet aperçu, nous voyons que les données de références sont groupées en sept colonnes: quatre colonnes contenant des variables continues, deux colonnes contenant des variables catégoriques, dont la variable "type" représentant les cibles de notre modèle, et la première colonne contient l'indexe principal du tableau. Cette dernière pourra être ignorée lors de l'analyse. Finalement, nous remarquons que chaque ligne correspond à une seule observation et chaque colonne représente une seule variable, alors nous n'avons pas à modifier le tableau pour qu'il respecte les principes du "tidy data".

Pour explorer les données, j'ai créé un échantillon aléatoire contenant 70% des données que j'ai enregistrées dans le data frame nommé "train". Ce dernier nous servira de référence lors de notre analyse, alors que le 30%  restant est réservé dans le data frame "cross_val" que nous utiliserons lors de la validation croisée.

```r
train.indice <- sample(1:nrow(monstres), round(nrow(monstres)*0.7))
train <- monstres[train.indice , ]
cross_val <- monstres[-train.indice , ]
```

Imprimons les statistiques descriptives principales de chacune des variables:

```r
summary(monstres)
sum(is.na(monstres))
```

Bonne nouvelle! Il n'y a aucune donnée manquante dans le tableau, alors je vais pouvoir utiliser chaque observation sans avoir à estimer la valeur des données manquantes. Une autre catéchistique intéressante de ces données est que chaque variable continue a une valeur maximale et une valeur minimum approximativement égale à 1 et 0, respectivement. Cela nous assure qu'une variable n'aura pas d'influence indue sur les algorithmes que nous allons utiliser. Si cela n'avait pas été le cas, il aurait été préférable de faire une normalisation des données. 

En regardant les statistiques descriptives de plus près, nous voyons que chaque variable continue est distribuée selon une distribution ayant une moyenne et une médiane qui sont à peu près égales et leurs premiers et leurs troisièmes quantiles sont à distance sensiblement égale de la moyenne ce qui peut indiquer que ces variables sont distribuées selon une loi normale. Affichons les graphiques des distributions de ces variables continues et appliquons le test de Normalité de Shapiro pour vérifier cette hypothèse.

```r
ggplot(monstres,aes(bone_length))+geom_density(fill="blue")+ theme_bw()
```

```r
shapiro.test(monstres$bone_length)
```

```r
ggplot(monstres,aes(rotting_flesh))+geom_density(fill="blue")+ theme_bw()
```

```r
shapiro.test(monstres$rotting_flesh)
```

```r
ggplot(monstres,aes(hair_length))+geom_density(fill="blue")+ theme_bw()
```

```r
shapiro.test(monstres$hair_length)
```

```r
ggplot(monstres,aes(has_soul))+geom_density(fill="blue")+ theme_bw()
```

```r
shapiro.test(monstres$hair_length)
```

Ces graphiques viennent renforcer notre hypothèse selon laquelle ces variables sont distribuées selon des lois normales. De plus, les tests de normalité pour chacune des variables ne nous permettent pas de rejeter l'hypothèse selon laquelle ces variables sont distribuées selon une loi normale. Le graphique suivant montre la distribution des couleurs selon les étiquettes de nos observations.

```r
ggplot(monstres,aes(x = color,fill=type,color=type))+
  stat_count()+ggtitle("Répartition des couleurs selon le type de monstre")
```

Ce graphique montre clairement que dans chaque catégorie de la variable "color", la proportion d'observations de chacun des trois types de monstre est sensiblement identique. Vérifions cette hypothèse en plaçant ces données dans un tableau et en calculant le pourcentage d'écart entre la moyenne et la fréquence de chacun des types d'observation.

```r
proportion<-data.frame(table(monstres$color,monstres$type))
proportion<-dcast(proportion, Var1~Var2)
```

```r
transform(proportion, Moyenne=round(rowMeans(proportion[,2:4]),2), 
          Ecart.Ghost=round(1-(proportion[,2])/rowMeans(proportion[,2:4]),2)*100,
         Ecart.Ghoul=round(1-(proportion[,3])/rowMeans(proportion[,2:4]),2)*100,
          Ecart.Goblin=round(1-(proportion[,4])/rowMeans(proportion[,2:4]),2)*100)
```

Nous voyons que pour la plupart des couleurs, le pourcentage d'écart est sensiblement égal entre les trois catégories. Cela nous indique que cette variable ne permet pas de discerner entre chacune des trois catégories et donc qu'elle n'aura pas beaucoup d'influence positive sur le modèle.

Maintenant, mesurons la corrélation entre les variables continues du modèle pour vérifier si certaines d'entre elles devraient être omises du modèle. Nous allons dessiner une matrice de corrélation pour évaluer visuellement s’il existe des liens entre ces quatre variables.

```r
corrplot.mixed(cor(monstres[,2:5,with=F]), lower="ellipse", upper="number")
```

Cette matrice de corrélation montre qu'il existe une corrélation positive relativement forte entre les variables "hair_length" et "has_soul", ainsi que de faibles corrélations positivent entre les variables "bone_length/hair_length" et "bone_length/has_soul". Ces corrélations ne sont pas assez fortes pour justifier l'élimination d'une variable du modèle.

Les nuages de points ci-dessous montrent comment sont dispersés les types de montres selon chaque paire de variables.


```r
pairs(subset(train, select = c("bone_length","rotting_flesh","hair_length","has_soul")),
             main = "Répartition des données pour chaque paire de variable",
      pch = 21, bg = c("red2", "green3", "blue")[unclass(train$type)])
```

Ces nuages de points montrent qu'il n'existe pas de frontière claire entre les données de chacune des trois catégories d'observations. Puisque l'efficacité des algorithmes de partitions diminue considérablement lorsque les données ne sont pas séparables, on en conclut qu'il ne sera pas trivial d'obtenir un bon taux de classification en utilisant uniquement les variables précédentes.

# Nouvelles variables

Maintenant, essayons de créer de nouvelles variables qui contribueront à améliorer notre modèle. Puisque nos données ne sont pas linéairement séparables, nous voulons créer des variables qui vont distancer les trois catégories. Regardons la distribution de la variable "has_soul" selon le type de monstre.

```r
qplot(type,has_soul, data=monstres, geom=c("boxplot", "jitter"),
   fill=type, main="has_soul en fonction du type de monstre", xlab="", ylab="has_soul") 
```

```r
mean(subset(monstres, type == "Ghost")$has_soul)
mean(subset(monstres, type == "Ghoul")$has_soul)
mean(subset(monstres, type == "Goblin")$has_soul)
```

Nous voyons que les moyennes des données provenant des catégories Ghost et Ghoul sont distinctes. En multipliant cette variable par elle-même, nous augmentons la distance entre les points de ces deux catégories. Pour ce faire, nous créons la variable soul_square et l'ajoutons à notre data frame monstres.

```r
monstres$soul_square=monstres$has_soul*monstres$has_soul
qplot(type,soul_square, data=monstres, geom=c("boxplot", "jitter"),
   fill=type, main="soul_square en fonction du type de monstre", xlab="", ylab="soul_square")
```

```r
mean(subset(monstres, type == "Ghost")$soul_square)
mean(subset(monstres, type == "Ghoul")$soul_square)
mean(subset(monstres, type == "Goblin")$soul_square)
```

Les résultats précédents montrent que la multiplication de la variable has_soul déplace les trois distributions vers 0, mais que la moyenne du nuage de point de la variable Ghoul est maintenant trois fois plus grande que la moyenne du nuage de point de la variable Ghost. Cela en soi est un indicateur que cette variable pourra contribuer positivement au modèle. Malheureusement, l'opération à aussi étendu le nuage de point de la variable Goblin et cela va ajouter beaucoup de bruit dans les observations, ce qui pourra contrer les avantages de l'opération.   

Regardons la distribution des autres variables pour voir si le même raisonnement peut s'appliquer à celles-ci.

```r
qplot(type,hair_length, data=monstres, geom=c("boxplot", "jitter"),
   fill=type, main="hair_length en fonction du type de monstre", xlab="", ylab="hair_length") 
mean(subset(monstres, type == "Ghost")$hair_length)
mean(subset(monstres, type == "Ghoul")$hair_length)
mean(subset(monstres, type == "Goblin")$hair_length)
```

```r
qplot(type,bone_length, data=monstres, geom=c("boxplot", "jitter"),
   fill=type, main="bone_length selon le type de monstre", xlab="", ylab="bone_length") 
mean(subset(monstres, type == "Ghost")$bone_length)
mean(subset(monstres, type == "Ghoul")$bone_length)
mean(subset(monstres, type == "Goblin")$bone_length)
```

```r
qplot(type,rotting_flesh, data=monstres, geom=c("boxplot", "jitter"),
   fill=type, main="bone_length", xlab="", ylab="OS") 
mean(subset(monstres, type == "Ghost")$rotting_flesh)
mean(subset(monstres, type == "Ghoul")$rotting_flesh)
mean(subset(monstres, type == "Goblin")$rotting_flesh)
```

Nous voyons que le même raisonnement peut s'appliquer aux variables "bone_length" et "hair_length", mais que pour la variable "rotting_flesh" la différence entre les moyennes des nuages de points n'est pas assez grande pour que la multiplication sépare les points des trois catégories. De plus, nous voyons que la position relative des nuages de points dans ce dernier graphique est particulière. En effet, c'est seulement pour la variable "rotting_flesh" que la moyenne des données provenant de la catégorie "Ghost" est supérieure à la moyenne des autres catégories. Cela nous indique que cette variable nous permettra de différencier entre les catégories "Ghost" et "Goblin" et aura une influence positive sur le modèle.

```r
monstres$hair_square=monstres$hair_length*monstres$hair_length
monstres$soul_square=monstres$has_soul*monstres$has_soul
monstres$bone_square=monstres$bone_length*monstres$bone_length
```

Nous pouvons étendre ce raisonnement aux paires de variables. En effet, les distributions des variables "hair_length" et "has_soul" ont sensiblement la même séparation entre les nuages de points des Ghost et des Goblins, donc multiplier ces deux variables nous permettra d'obtenir une nouvelle variable dont les données sont plus séparées.

```r
monstres$hair_time_soul=monstres$hair_length*monstres$has_soul
qplot(type,hair_time_soul, data=monstres, geom=c("boxplot", "jitter"),
   fill=type, main="hair_time_soul", xlab="", ylab="hair_time_soul") 
```

```r
mean(subset(monstres, type == "Ghost")$hair_time_soul)
mean(subset(monstres, type == "Ghoul")$hair_time_soul)
mean(subset(monstres, type == "Goblin")$hair_time_soul)
```

La présence de corrélation entre les variables "hair_length", "has_soul" et "bone_length", ainsi que les nuages de points nous indique quelles sont de bonnes candidates pour la création de nouvelles variables.

```r
monstres$hair_time_soul=monstres$hair_length*monstres$has_soul
monstres$hair_time_bone=monstres$hair_length*monstres$bone_length
monstres$bone_time_soul=monstres$bone_length*monstres$has_soul
```
