```python
import os
import re
import csv
import json
import locale
import collections

base_dir = "E:\\Users\Guillaume\Downloads\movie_rating_prediction-master\movie_rating_prediction-master"
#locale.setlocale( locale.LC_ALL, 'en_US.UTF-8' )

def parse_year(year):
    if not year:
        return 0
    return locale.atoi(year[-4:])

def parse_price(price):
    if not price:
        return 0
    return locale.atoi(re.sub('[^0-9]', "", price))

def load_unparsed_movie_metadata():
    try:
        with open(os.path.join(base_dir, "movie_budget.json"), "r") as f:
            movies = json.load(f)
            return movies
    except:
        print ("Cannot load the unparsed movie metadata file!")
        return None
    
def parse_one_movie_metadata(movie):
    if not movie:
        return None

    parsed_movie = {}

    parsed_movie['budget'] = None if movie['production_budget'] is None or len(movie['production_budget']) == 0 else parse_price(movie['production_budget'].strip())
    parsed_movie['gross'] = None if movie['domestic_gross'] is None or len(movie['domestic_gross']) == 0 else parse_price(movie['domestic_gross'].strip())
    parsed_movie['title_year'] = None if movie['release_date'] is None else int(parse_year(movie['release_date']))
    parsed_movie['movie_title'] = movie['movie_name'].encode('utf-8')

    return parsed_movie

def parse_all_budget():
    movies = load_unparsed_movie_metadata()
    total_count = len(movies)
    print ("{} movie metadata were loaded!".format(total_count))

    with open("scrap_gross_budget.csv", "w") as f:
        header_was_written = False
        for i, movie in enumerate(movies):
            print ("Processing {} of {}: {}".format(i+1, total_count, movie['movie_name']))
            parsed_movie = parse_one_movie_metadata(movie)
            w = csv.DictWriter(f, parsed_movie.keys())
            if not header_was_written:
                w.writeheader()
                header_was_written = True

            w.writerow(parsed_movie)
            
def compare_movie_title():
    movies = load_unparsed_movie_metadata()
    budget = load_unparsed_movie_metadata()
    
    

```

```python
import difflib


base_dir = "E:\\Users\Guillaume\Downloads\movie_rating_prediction-master\movie_rating_prediction-master"


def load_fetch_imdb_url():
    try:
        with open(os.path.join(base_dir, "fetch_imdb_url.json"), "r") as f:
            movies = json.load(f)
            return movies
    except:
        print ("Cannot load the unparsed movie metadata file!")
        return None

def count_diff(name_a, name_b):
    count_length=0
    count_diff=0
    for i,s in enumerate(difflib.ndiff(name_a, name_b)):
            count_length=count_length+1
            if s[0]==' ': continue
            elif s[0]=='-':
                count_diff=count_diff+1
            elif s[0]=='+':
                count_diff=count_diff+1
    return(count_diff/count_length)

def parse_name_IMDB(movie):
    if not movie:
        return None

    parsed_movie = {}
    parsed_movie['movie_name_IMDB'] = movie['movie_name'].encode('utf-8')

    return parsed_movie

def parse_name_budget(movie):
    if not movie:
        return None

    parsed_movie = {}
    parsed_movie['movie_name_budget'] = movie['movie_name'].encode('utf-8')

    return parsed_movie
    
def compare_movie_title():
    names_budget = load_unparsed_movie_metadata()
    names_IMDB =load_fetch_imdb_url()
    total_count = len(names_budget)
    total_count_IMDB = len(names_IMDB)
    print ("{} movie in budget".format(total_count))
    print ("{} movie in IMDB".format(total_count_IMDB))

    with open("name_comparason.csv", "w") as f:
        header_was_written = False
        for i, movie in enumerate(names_budget):
            parsed_budget = {}
            parsed_budget['movie_name_budget'] = movie['movie_name']
            #parsed_budget = parse_name_budget(movie)
            print ("Processing {} of {}: {}".format(i+1, total_count, movie['movie_name']))
            min_diff=100
            simil_name=""
            for j, name in enumerate(names_IMDB):
                    temp=count_diff(parsed_budget['movie_name_budget'],name['movie_name'])
                    if temp<min_diff:
                        min_diff=temp
                        simil_name=name['movie_name']

            parsed_budget['movie_name_IMDB']=simil_name    
            w = csv.DictWriter(f, parsed_budget.keys())
            if not header_was_written:
                w.writeheader()
                header_was_written = True

            w.writerow(parsed_budget)
                
   
                   
            
```
